```python
# Lá»c cÃ¡c quá»‘c gia cÃ³ Series Name vÃ  giÃ¡ trá»‹ lá»›n hÆ¡n value
def filter_data_by_series_code(series_name, value):
    return data[data[series_name] > value] if len(data[data[series_name] > value]) > 0 else 'KhÃ´ng cÃ³ nÆ°á»›c nÃ o'
```

```python
interact(filter_data_by_series_code, series_name=data.columns[2:], value=1000)
```

```python
# Giá»›i thiá»‡u sÆ¡ qua vá» dá»¯ liá»‡u cÃ¡c cá»™t
data.hist(figsize=(30, 25), bins=50, xlabelsize=8, ylabelsize=8, color='red', alpha=0.4)
```

```python
# Sáº¯p xáº¿p tÄƒng dáº§n theo Series Name
def sort_by_series_name(series_name):
    return data[data.columns[2:]].sort_values(by=series_name)
interact(sort_by_series_name, series_name=data.columns[2:])
```

```python
# HÃ m tÃ¬m quá»‘c gia vá»›i series vÃ  giÃ¡ trá»‹ max, min
def find_country(operation, selected_serie_name):
    # Náº¿u chá»n giÃ¡ trá»‹ lá»›n nháº¥t
    if operation == 'Max':
        # TÃ¬m giÃ¡ trá»‹ lá»›n nháº¥t
        max_value = data[selected_serie_name].max()
        # Hiá»ƒn thá»‹ táº¥t cáº£ cÃ¡c quá»‘c gia cÃ³ giÃ¡ trá»‹ báº±ng giÃ¡ trá»‹ lá»›n nháº¥t
        print(data[(data[selected_serie_name] == max_value)]['Country Name'].unique() + ' cÃ³ giÃ¡ trá»‹ lá»›n nháº¥t lÃ : ' + str(max_value))
    # Náº¿u chá»n giÃ¡ trá»‹ nhá» nháº¥t
    elif operation == 'Min':
        # TÃ¬m giÃ¡ trá»‹ nhá» nháº¥t
        min_value = data[selected_serie_name].min()
        # Hiá»ƒn thá»‹ táº¥t cáº£ cÃ¡c quá»‘c gia cÃ³ giÃ¡ trá»‹ báº±ng giÃ¡ trá»‹ nhá» nháº¥t
        print(data[(data[selected_serie_name] == min_value)]['Country Name'].unique() + ' cÃ³ giÃ¡ trá»‹ nhá» nháº¥t lÃ : ' + str(min_value))
```

```python
# CÃ¡c series lÃ  báº¯t Ä‘áº§u tá»« cá»™t thá»© 3
list_series = data.columns[2:]
interact(find_country, operation=[
         'Max', 'Min'], selected_serie_name=list_series)
```

```python
def seaborn_chart(selected_serie_name, selected_chart_type, selected_year):
    # Láº¥y ra dá»¯ liá»‡u tá»« nÄƒm tÆ°Æ¡ng á»©ng
    selected_data = data[data['Year'] == selected_year]
    # Liá»ƒm tra cÃ³ serie name trong cá»™t dá»¯ liá»‡u khÃ´ng
    if selected_serie_name in selected_data.columns:
        # Váº½ biá»ƒu Ä‘á»“
        plt.figure(figsize=(12, 12))
        # Kiá»ƒm tra xem dáº¡ng biá»ƒu Ä‘á»“ nÃ o
        if selected_chart_type == 'Line Chart':
            sns.lineplot(x='Country Name', y=selected_serie_name, data=selected_data)
        elif selected_chart_type == 'Bar Chart':
            sns.barplot(x='Country Name', y=selected_serie_name, data=selected_data)
        elif selected_chart_type == 'Pie Chart':
            plt.pie(selected_data[selected_serie_name], labels=selected_data['Country Name'], autopct='%1.1f%%', shadow=True)
        elif selected_chart_type == 'Scatter Plot':
            sns.scatterplot(x='Country Name', y=selected_serie_name, data=selected_data)
        elif selected_chart_type == 'Box Plot':
            sns.boxplot(x='Country Name', y=selected_serie_name, data=selected_data)
        elif selected_chart_type == 'Violin Plot':
            sns.violinplot(x='Country Name', y=selected_serie_name, data=selected_data)
        elif selected_chart_type == 'Cat Plot':
            sns.catplot(x='Country Name', y=selected_serie_name,kind='box', data=selected_data)
        # TÃªn biá»ƒu Ä‘á»“
        plt.title(f'{selected_serie_name} in {selected_year}')
        plt.xlabel('Country Name')
        plt.ylabel(selected_serie_name)
        plt.grid(True)
        plt.show()
    else:
        print(f"Selected series '{selected_serie_name}' not found in the dataset for the selected year.")
# LiÃªn káº¿t hÃ m váº½ biá»ƒu Ä‘á»“ vá»›i sá»± kiá»‡n chá»n serie name, loáº¡i biá»ƒu Ä‘á»“, vÃ  nÄƒm
interact(seaborn_chart, selected_serie_name=data.columns[2:], selected_chart_type=[
         'Line Chart', 'Bar Chart', 'Pie Chart', 'Scatter Plot', 'Box Plot', 'Violin Plot', 'Cat Plot'], selected_year=data['Year'].unique())
```
