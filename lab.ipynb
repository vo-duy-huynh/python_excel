```python
# Lọc các quốc gia có giá trị "Average precipitation in depth (mm per year)" lớn hơn 1000 theo năm
def filter_and_print( year_column):
    actual_year = year_column + ' [YR' + year_column + ']'
    filtered_countries = data[data['Series Code'] == 'AG.LND.PRCP.MM']
    filtered_countries = filtered_countries[filtered_countries[actual_year] > 1000]
    # Chọn các cột cần hiển thị
    display_columns = ['Country Name', 'Country Code', 'Series Name', 'Series Code', actual_year]
    # Hiển thị kết quả mà không bị xuống dòng
    print(filtered_countries[display_columns].to_string(index=False))
```

```python
# Các năm là bắt đầu từ cột thứ 4
list_years = data.columns[4:]
# Xử lý tên cột để lấy năm
list_years = list_years.str.split(' ').str[0]
interact(filter_and_print, year_column=list_years)
```

```python
# Tạo hàm vẽ biểu đồ bằng matplot
def plot_chart(selected_country, selected_chart_type, selected_series_code):
    # Xoá biểu đồ cũ đi
    plt.clf()
    # Chọn dòng có Country Name, Series Code và loại biểu đồ như đã chọn
    selected_data = data[(data['Country Name'] == selected_country) &
                         (data['Series Code'] == selected_series_code)]

    # Xóa cột không cần thiết
    selected_data = selected_data.drop(
        ['Country Name', 'Country Code', 'Series Name', 'Series Code'], axis=1)
    # Lấy ra dòng tên cột cụ thể là các năm
    selected_data_long = selected_data.iloc[0]
    # Xử lý chuyển chuỗi để lấy ra năm tương ứng,
    selected_data_long.index = selected_data_long.index.str.split(' ').str[0]
    selected_data_long = selected_data_long.reset_index()
    selected_data_long.columns = ['Year', 'Value']

    # Vẽ biểu đồ
    plt.figure(figsize=(10, 6))
    if selected_chart_type == 'Line Chart':
        plt.plot(selected_data_long['Year'], selected_data_long['Value'], marker='o')
    elif selected_chart_type == 'Bar Chart':
        plt.bar(selected_data_long['Year'], selected_data_long['Value'])
    plt.title(f'{selected_country} - {selected_series_code}')
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.grid(True)
    plt.show()
```

```python
# Liên kết hàm vẽ biểu đồ với sự kiện chọn quốc gia, loại biểu đồ, và Series Code
interact(plot_chart, selected_country=data['Country Name'].unique(),
         selected_chart_type=['Line Chart', 'Bar Chart'], selected_series_code=data['Series Code'].unique())
```

```python
# Hàm tìm quốc gia với năm và giá trị
def find_country(operation, selected_year):
    # actual_year là tên cột có năm được chọn kết hợp với [YR...]
    actual_year = selected_year + ' [YR' + selected_year + ']'
    # Nếu chọn tất cả các năm
    if operation == 'All':
        # Hiển thị tất cả các quốc gia có giá trị khác 0
        print(data[data[actual_year] != 0]['Country Name'].unique())
    # Nếu chọn giá trị lớn nhất
    elif operation == 'Max':
        # Tìm giá trị lớn nhất
        max_value = data[actual_year].max()
        # Hiển thị tất cả các quốc gia có giá trị bằng giá trị lớn nhất
        print(data[(data[actual_year] == max_value)]['Country Name'].unique() + ' có giá trị lớn nhất là: ' + str(max_value))
    # Nếu chọn giá trị nhỏ nhất
    elif operation == 'Min':
        # Tìm giá trị nhỏ nhất
        min_value = data[actual_year].min()
        # Hiển thị tất cả các quốc gia có giá trị bằng giá trị nhỏ nhất
        print(data[(data[actual_year] == min_value)]['Country Name'].unique() + ' có giá trị nhỏ nhất là: ' + str(min_value))
```

```python
# Các năm là bắt đầu từ cột thứ 4
list_years = data.columns[4:]
# Xử lý tên cột để lấy năm
list_years = list_years.str.split(' ').str[0]
interact(find_country, operation=[
         'All', 'Max', 'Min'], selected_year=list_years)
```

```python
# Biểu đồ đường và cột bằng seaborn
def seaborn_chart(selected_country, selected_chart_type, selected_series_code):
    # Xoá biểu đồ cũ đi
    plt.clf()
    # Chọn dòng có Country Name, Series Code và loại biểu đồ như đã chọn
    selected_data = data[(data['Country Name'] == selected_country) &
                         (data['Series Code'] == selected_series_code)]

    # Xóa cột không cần thiết
    selected_data = selected_data.drop(
        ['Country Name', 'Country Code', 'Series Name', 'Series Code'], axis=1)
    selected_data_long = selected_data.iloc[0]
    selected_data_long.index = selected_data_long.index.str.split(' ').str[0]
    selected_data_long = selected_data_long.reset_index()
    selected_data_long.columns = ['Year', 'Value']

    # Vẽ biểu đồ
    plt.figure(figsize=(12, 6))

    if selected_chart_type == 'Line Chart':
        sns.lineplot(x='Year', y='Value', data=selected_data_long, marker='o')
    elif selected_chart_type == 'Bar Chart':
        sns.barplot(x='Year', y='Value', data=selected_data_long)
    #Biểu đồ tròn
    # elif selected_chart_type == 'Pie Chart':
    #     plt.pie(selected_data_long['Value'], labels=selected_data_long['Year'], autopct='%1.1f%%', startangle=140)
    #     plt.title(f'{selected_country} - {selected_series_code} - Pie Chart')
    #     plt.show()

    plt.title(f'{selected_country} - {selected_series_code}')
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.grid(True)
    plt.show()
```

```python
# Liên kết hàm vẽ biểu đồ với sự kiện chọn quốc gia, loại biểu đồ, và Series Code
interact(seaborn_chart, selected_country=data['Country Name'].unique(),
         selected_chart_type=['Line Chart', 'Bar Chart'],
         selected_series_code=data['Series Code'].unique())
```
